```python
# import packages
import pandas as pd
import numpy as np
import xlsxwriter

# Read Initial file
df = pd.read_excel('text_analytics.xlsx')
df.info()
```

```python
# Find Story Length - COUNTING ALL CHARACTERS
df['Story_Length']= df['Project_Story'].str.len() 
# Find Story number of words - COUNTING ONLY THE NUMBER OF WORDS
df['wordcount_story'] = df['Project_Story'].apply(lambda x: len(str(x).split(' ')))
df
```

```python
import nltk
nltk.download('stopwords')
stop = stopwords.words('english')

# Number of english stopwords
df['stopwords_story_count'] = df['Project_Story'].apply(lambda x: len([x for x in str(x).split(' ') if x in stop]))
```

```python
# Boolean variable of title being the same
df['Same_Title'] = df['Project_Title_Scrap'] == df['Project_Title']
```

```python
# Calculate Similarity between the strings of the titles
import difflib as diff
df['similarity_titles'] = df.apply(lambda x: diff.SequenceMatcher(None, str(x['Project_Title_Scrap']).strip(), str(x['Project_Title']).strip()).ratio(), axis=1)
df['similarity_titles'] = df['similarity_titles']*100
df
```

```python
# discover textstat

# Reading Ease Score
# import textstat
# df['reading_ease_score'] = df.apply(lambda row: textstat.flesch_reading_ease(str(row['Project_Story'])), axis=1)
```

```python
# Count spaces within the story
df['spaces_story_count'] = df['Project_Story'].apply(lambda x: str(x).count(' '))
```

```python
# Count number of sentences through finding sum of: '.', '!', '?'
df['sentence_count'] = df['Project_Story'].apply(lambda x: str(x).count('.') + str(x).count('!') + str(x).count('?'))
df
```

```python
# Start text analytics

from nltk import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
from nltk.stem import LancasterStemmer, WordNetLemmatizer, PorterStemmer
from wordcloud import WordCloud, STOPWORDS
from textblob import TextBlob
```

```python
def senti(x):
    return TextBlob(x).sentiment
```

```python
df['Project_Story'] = df['Project_Story'].astype(str)
df['sentiment_score'] = df['Project_Story'].apply(senti)
```

```python
writer = pd.ExcelWriter('first_try_text_analytics.xlsx', engine='xlsxwriter')

df.to_excel(writer)

# Output the Excel file.
writer.save()
```
